```python
def calculate_life_quartile(exp):
    
    """ This function defines the life expectancy quartile'
        input: exp is the life expectancy in years
        returns: an integer representing the quartile 
    """
                
    if exp < 58.41:
        # This observation is in the first quartile    -> keep in mind no min age was defined here so a negative number would be valid
        return 1
    elif exp >= 58.41 and exp < 67.05:
        # this observation is the second quartile
        return 2
    elif exp >= 67.05 and exp < 71.70:
        # This observation is in the third quartile
        return 3
    elif exp >= 71.70:
        # This observation is in the fourth quartile
        return 4
    else:
        # We might have bad data
        return None
        
```

```python
calculate_life_quartile(80.6)
```

```python
import pandas as pd
```

```python
data = pd.read_csv('data/gapminder_all.csv')
data['life_qrtl'] = data['lifeExp_1987'].apply(calculate_life_quartile)
```
